### Arrays

Arrays simply store many varibles. They can store different datatypes but not together e.g. you could have an int array or a float array but they should be made of all ints or all floats.

Arrays are defined with square brackets.

    int myIntArray[] = {0,1,2,3}
    
    float myFloatArray[] = {0.01, 0, 0.3}

We can also leave them empty originally and just define their size. For example:

```c
#include <stdio.h>

int main() {

    int numbers[2];
    double marks[7];
    float interest[5];

}
```

When we do this we need to know the size of the arrays - the program then allocates enough memory for this array. Once you have created the array you canot change its size.

### Functions

In C the syntax for creating a function looks like this;

    void myFunction(comma separated parameters with their datatype) {
      // code here
    }
    
If void is used as above then it means that the function you are creating does not have a value which it returns. 

Functions can be created outside of main and then stored up for future use. E.g.

```c
#include <stdio.h>

void myFirstFunction() {
        printf("This is my first function");
}

int main() {

    myFirstFunction();
    
    return 0;
}
```

```c
#include <stdio.h>

void myFirstFunction(char strValue[]) {
        printf("This is my %s function \n", strValue);
}

int main() {

    char strValue[] = "second";
    myFirstFunction(strValue);
    
    return 0;
}
```

#### Recursion

Recursion is essentially writing a function which can call itself. It is a very useful technique as it allows you to simplfy a task. 

For exmaple we could think aboutt he more complicated task of adding a series of numbers toegther as the repition of a simpler task of adding two numbers togther. 

Let's look at an example of this below:

    int sum(int k) {
      if (k > 0) {
        return k + sum(k - 1);
      } else {
        return 0;
      }
    }

This function takes the integer that you give it and says:

    if that integer > 0 then 
        return that integer + the result of the fucntion called for te integer -1
        This will reccur untill k=0
    at k=0 return 0 (this breaks the loop)
    
Let's see that fucntion called within a C program WITH declaration and definition separated. 

```c
#include <stdio.h>

int sum(int k);

int main() {
  int result = sum(3);
  printf("%d", result);
  return 0;
}

int sum(int k) {
  if (k > 0) {
    return k + sum(k - 1);
  } else {
    return 0;
  }
}
```

Q: write a recursive function which takes a number n and returns the nth number of the fibonacci series.

```c
#include <stdio.h>

int fib(int k);

int main() {
  int result = fib(6);
  printf("%d", result);
  return 0;
}

int fib(int k) {
  if (k == 0 | k == 1) {
    return k;
  } else {
    return fib(k-1) + fib(k-2);
  }
}
```

### Enumerations

An enumeration assigns integer constants to variables. A common use case to see why this might be hepful is the days of the week or the months of the year. Mon through Sunday can be assigned an integer number 0 through 6. 

```c
#include<stdio.h>
 
enum week{Mon, Tue, Wed, Thur, Fri, Sat, Sun};
 
int main()
{
    enum week day;
    day = Sun;
    printf("%d",day);
    return 0;
}
```

We can also assign the integer values ourselves using

    enum name{another_name = 2, yet_another_name=0};
    
although the above use case where we assign 'flags' in order is perhaps more common.

### Pointers

Variables are always assigned a location and amount of memory. In C the programmer makes these memory choices while it was mainly hidden from you in python.

Every variable in C has a 'pointer' which stores the address of where that memory was stored. 

Pointers are assigned with a * and referenced with an & so in the below &myFloat is a pointer for the variable myFloat.

```c
#include <stdio.h>

int main() {

    float myFloat = 1.1;
    
    printf("%p", &myFloat);
    
    return 0;
}
```

This jumble of stuff is a hexadecimal address which represents where the variable 'myFloat' is now stored.

So let's see how the pointer is assigned to a variable.

```c
#include <stdio.h>

int main() {

    float myFloat = 1.1;
    float* myFloatPtr = &myFloat;
    
    printf("%p\n", &myFloat);
    printf("%p\n", myFloatPtr);

    printf("%d\n", myFloat);
    printf("%d", *myFloatPtr);

    return 0;
}
```

So you can see there are two ways to access the varible and it's address.

### Structures

A structure is a type of derived datatype which can store many different types of datatypes within a memory block. It will allocate the memory for each of these together and store them in a single place. 

    struct birthday {
        char name[20];
        int day;

        int year;
    };


### Unions

A union is very similar to a struct except it can only store one set of the data that it has at a time. This means if you try to access all the data at once it will not be correct. Only the last thing you stored in that memory location would be correct and the rest would be corupted. See the two examples below:

```c
#include <stdio.h>
#include <string.h> // this library included for builtin use of strcpy

/* here we are calling the id 1, group a, and name student1
- do we get them correct? */

int main() {

  union Student{
   char name[25];
   int id;
   char group;
   float marks[5];
   double interest;
   }st1, st2;
 
  st1.id = 1; 
  st1.group = 'a';
  strcpy(st1.name, "student1");
  printf( "ID : %d\n", st1.id);
  printf( "Group : %c\n", st1.group);
  printf( "Name : %s\n", st1.name);

  return 0;
}
```

```c
#include <stdio.h>
#include <string.h>

/* let's try assigning them and calling them one-by-one
- are they correct this time? */

int main() {

  union Student{
    char name[25];
    int id;
    char group;
    float marks[5];
    double interest;
    }st1, st2;
 
  st1.id = 1; 
  printf( "ID : %d\n", st1.id);
  st1.group = 'a';
  printf( "Group : %c\n", st1.group);
  strcpy(st1.name, "student1");
  printf( "Name : %s\n", st1.name);

  return 0;
}
```
